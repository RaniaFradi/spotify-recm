```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("spotify.csv")
data.head(5)
```

```python
data.isnull().sum()
```

```python
from sklearn.preprocessing import MinMaxScaler

dataTypes = ['int16', 'int32', 'int64', 'float32', 'float64']
normalization = data.select_dtypes(include = dataTypes)#all the numerical columns
for cul in normalization.columns:
    MinMaxScaler(cul)
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans()
data['features'] = kmeans.fit_predict(normalization)

MinMaxScaler(data['features'])
```

```python
from tqdm import tqdm #to show precess

class Spotify_Recommendation_System():
    
    def __init__(self, dataset):
        
        self.dataset = dataset
        
    def recommendation(self, songs):
        
        distance = []
        song = self.dataset[(self.dataset.name.str.lower() == songs.lower())].head(1).values[0]
        recm = self.dataset[self.dataset.name.str.lower() != songs.lower()]
        
        for s in tqdm(recm.values):
            dist = 0
            
            for col in np.arange(len(recm.columns)):
                if col not in [1, 6, 12, 14, 18]: #not acousticness, explicit, mode, popularity or valence
                    dist += np.absolute( float(song[col])-float(s[col]) )
                    
            distance.append(dist)
            
        recm['distance'] = distance
        recm = recm.sort_values('distance')
        columns = ['id', 'artists', 'name']
        
        return recm[columns][:10]

recommendations = Spotify_Recommendation_System(data)
recommendations.recommendation("Xuniverxe")
```
